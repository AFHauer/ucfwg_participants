```python
import pandas as pd
import matplotlib.pyplot as plt

part = pd.read_csv('participants.csv')

part.head(5)
```

```python
mean_part = round(part['participants'].mean())

mean_part
```

```python
part.plot.bar(y='participants', x='month', 
              xlabel='Month', ylabel='Participants', 
              title='UCFWG Discussion Topic Meeting Participants', 
              legend=False, figsize=(12,8))

```

```python
x = part['month']
y = part['participants']
t = part['topic']

fig, ax = plt.subplots(figsize=(13,10))
bars = ax.bar(x, y, width=0.6, 
              color="slategray")

for bar in bars:
    height = bar.get_height()
    label_x_pos = bar.get_x() + bar.get_width() / 2
    ax.text(label_x_pos, height, s=f'{height}', ha='center',
            va='bottom', size=14, color="navy")


plt.title('UCFWG Discussion Topic Meeting Participants \n Average of {} participants'.format(mean_part), size=18)
plt.xlabel('Month', size=15)
plt.ylabel('Participants', size=15)
plt.ylim([0,85])
plt.savefig('participants.png')

plt.show()

```
